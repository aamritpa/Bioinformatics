```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('output.xlsx')
```

```python
plt.plot(df['average_Intrabond'])
plt.xlabel('Number of Entries')
plt.ylabel('Average length')
plt.show()
```

```python
plt.plot(df['avg_Inside_Pairs_Length'])
plt.xlabel('Number of Entries')
plt.ylabel('Average length')
plt.show()
```

```python
plt.scatter(df['average_Intrabond'],df['avg_Inside_Pairs_Length'],c='blue')
```

```python
#Statistical Tests
ttest = stats.ttest_ind(df['average_Intrabond'], df['avg_Inside_Pairs_Length'])
print(ttest)
```

```python
#Correlation Coeeficiant
print(stats.linregress(df['average_Intrabond'],df['avg_Inside_Pairs_Length']).rvalue)
```

```python
#CHECKING MEAN
print(df['average_Intrabond'].mean(axis=0))
print(df['avg_Inside_Pairs_Length'].mean(axis=0))
```

```python
#EQUAL VARIANCE TEST
stats.levene(df['average_Intrabond'],df['avg_Inside_Pairs_Length']).pvalue
```
