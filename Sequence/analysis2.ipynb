```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data= pd.read_excel('sequenceOutput2.xlsx')
```

```python
newframe=pd.DataFrame()
newframe['Index']=pd.Series(['Sequence N_X_S/T count','Length','Sequence C count'])
```

```python
# cases
# Case1 No condition 
# Case2 w sequon and cystine >0, >0
# Case3 w sequon no cysteine Sequence C count=0 , NXST>0
# Case4 no sequon has cystein Sequence C count>0, NXST=0
# Case5 no cysteine and no sequon both ==0
```

```python
#Case 1 All data No restriction
mean1=np.mean(data['Sequence N_X_S/T count'])
mean2=np.mean(data['Length'])
mean3=np.mean(data['Sequence C count'])
newframe['No restriction(All Data)']= pd.Series([mean1,mean2,mean3])
```

```python
# Case2 w sequon and cystine >0, >0
df=data[(data['Sequence C count']>=1) & (data['Sequence N_X_S/T count']>=1) ]
mean1=np.mean(df['Sequence N_X_S/T count'])
mean2=np.mean(df['Length'])
mean3=np.mean(df['Sequence C count'])
newframe['sequon>0 and cystine >0']= pd.Series([mean1,mean2,mean3])
```

```python
# Case3 w sequon no cysteine Sequence C count=0 , NXST>0
df1=data[(data['Sequence C count']==0) & (data['Sequence N_X_S/T count']>=1) ]
mean1=np.mean(df1['Sequence N_X_S/T count'])
mean2=np.mean(df1['Length'])
mean3=np.mean(df1['Sequence C count'])
newframe['sequon>0 and cysteine=0']= pd.Series([mean1,mean2,mean3])
```

```python
# Case4 no sequon has cystein Sequence C count>0, NXST=0
df2=data[(data['Sequence C count']>=1) & (data['Sequence N_X_S/T count']==0) ]
mean1=np.mean(df2['Sequence N_X_S/T count'])
mean2=np.mean(df2['Length'])
mean3=np.mean(df2['Sequence C count'])
newframe['sequon=0 and cysteine>0']= pd.Series([mean1,mean2,mean3])
```

```python
# Case5 no cysteine and no sequon both ==0
df3=data[(data['Sequence C count']==0) & (data['Sequence N_X_S/T count']==0) ]
mean1=np.mean(df3['Sequence N_X_S/T count'])
mean2=np.mean(df3['Length'])
mean3=np.mean(df3['Sequence C count'])
newframe['sequon=0 and cysteine=0']= pd.Series([mean1,mean2,mean3])
```

```python
newframe=newframe.set_index('Index')
```
