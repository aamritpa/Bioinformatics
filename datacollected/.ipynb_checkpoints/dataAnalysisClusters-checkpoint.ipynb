```python
data1=pd.DataFrame()
data1['Term']=df1['Term']
data1['PValue1']=df1['PValue']
```

```python
data2=pd.DataFrame()
data2['Term']=df2['Term']
data2['PValue2']=df2['PValue']
```

```python
data3=pd.DataFrame()
data3['Term']=df3['Term']
data3['PValue3']=df3['PValue']
```

```python
data4=pd.DataFrame()
data4['Term']=df4['Term']
data4['PValue4']=df4['PValue']
```

```python
newdata=pd.merge(data1, data2, on='Term', how='right')
```

```python
newdata=pd.merge(data3,newdata, on='Term', how='right')
```

```python
newdata=pd.merge(data4,newdata, on='Term', how='right')
```

```python
newdata=newdata.fillna(0)
```

```python

plt.figure(figsize=(25,25))
plt.plot( newdata['PValue1'],newdata['Term'], 'bo',label='t1')
plt.plot( newdata['PValue2'],newdata['Term'], 'go',label='t2')
plt.plot( newdata['PValue3'],newdata['Term'], 'ro',label='t3')
plt.plot( newdata['PValue4'],newdata['Term'], 'yo',label='t4')

labels=['pval:Score=0','pval:Score>0_<=10','pval:Score>10_<=20','pval:Score>20']
plt.legend(labels)
plt.savefig('new.jpg')
```
